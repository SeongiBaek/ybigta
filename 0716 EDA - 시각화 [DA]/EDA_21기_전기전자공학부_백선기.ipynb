### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
df_missing = pd.DataFrame(df.isnull().sum(), columns=['missingValue'])
df_missing = df_missing[ df_missing['missingValue'] > 0 ]
df_missing = df_missing.sort_values(by ='missingValue', ascending = False)
df_missing['missingRatio'] = round(df_missing['missingValue'] / len(df) * 100, 2)

display(df_missing)
# import missingno as msno
# msno.matrix(df=df, figsize=(8, 8), color=(70/255, 172/255, 161/255))
```

```python
missingnum = df.isnull().sum()
missingnum = missingnum[missingnum>0]
q2df = pd.DataFrame(missingnum, columns=['missing_num'])
q2df = q2df.sort_values(by='missing_num', ascending = False)
q2df['missing_%'] = round(q2df['missing_num']/len(df)*100,1)
q2df
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
# sns.countplot("missingRatio", data = df_missing)
plt.figure(figsize = (50,16))

means = df_missing['missingRatio'].values.tolist()
labels = df_missing.index.values.tolist()

x = np.arange(len(df_missing))  
rec = plt.bar(x, means, 0.5) # 막대의 두께: 0.5
plt.xticks(x, labels)
plt.show()
```

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.
