#### merge 실습을 위한 데이터 쪼개기

```python
df_y = df_y[df_y.quality >= 7]
```

```python
len(df_y)
```

#### df_x와 df_y를 inner join시켜라 (quality를 기준으로)

```python
data_INNER_JOIN = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='inner')
display(data_INNER_JOIN)
```

#### df_x에서 df_y를 left join시켜라 (quality를 기준으로)

```python
data_LEFT_JOIN = pd.merge(df_x, df_y, left_on='quality', right_on='quality', how='left')
display(data_LEFT_JOIN)
```

- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
deliver = pd.read_csv('/content/gdrive/MyDrive/YBIGTA/deliver.csv') #wine.csv도 '내드라이브'에 사본 저장 후, 경로 수정
deliver.head()
```

```python
# deliver.loc[deliver['계절'][] > 10, '월'] = -500
def season(month):
    if month in ["03","04","05"]:
        return '봄'
    elif month in ["06","07","08"]:
        return '여름'
    elif month in ["09","10","11"]:
        return '가을'
    elif month in ["12","01","12"]:
        return '겨을'
    else:
        return '몰라'
```

```python
deliver['월']= deliver['날짜'].apply(lambda x: x[5:7])
deliver['계절']=deliver['월'].apply(season)
deliver.head(30)
```

```python
deliver['미세먼지(㎍/㎥)'].isnull().sum()
```

```python
deliver['미세먼지(㎍/㎥)'].fillna(deliver['미세먼지(㎍/㎥)'].mean(),inplace=True)
deliver.head(30)
```
